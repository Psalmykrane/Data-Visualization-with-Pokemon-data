```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pk = pd.read_csv("pokemon_data.csv")
```

```python
pk.info()
```

```python
pk.head()
```

```python
plt.hist(pk["Attack"])
plt.xlabel("Range of attacks")
plt.ylabel("Degree of attack")
plt.title("Distribution of Attack in pokemon data");
```

```python
plt.hist(pk["Sp. Atk"])
plt.xlabel("Range of special attacks")
plt.ylabel("Degree of special attack")
plt.title("Distribution of Special Attack in pokemon data");
```

```python
plt.boxplot(pk["Attack"], vert = False)
plt.xlabel("RANGE OF ATTACKS")
plt.title("Distribution of Attack in pokemon data");
```

```python
plt.boxplot(pk["Sp. Atk"], vert = False);
```

```python
pk.head()
```

```python
pk["Defense"].head(100).plot(
    kind = "line",
    xlabel = "Number of Defense",
    ylabel = "Degree of Defense",
    title = "Degree Data"
);
```

```python
sns.regplot(x = "Attack", y = "Defense", data = pk);
```

```python
sns.regplot(x = "Attack", y = "Sp. Atk", data = pk);
```

```python
correlation_matrix = pk.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation of Game features")
plt.xlabel("Skills")
plt.ylabel("Skills")
plt.show()
```

```python
sns.swarmplot(x='Attack', y='Defense', data=pk);
```
